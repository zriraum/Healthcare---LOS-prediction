## Training models using df with all classes as is

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
value = [ RandomForestClassifier(), DecisionTreeClassifier(),  
        KNeighborsClassifier(), CatBoostClassifier()]

key = [ 'RandomForsetClassifier', 'DecisionTreeClassifier',  
      'KNeighborsClassifier', 'CatBoostClassifier']

models= dict(zip(key,value))
accuracy_scores=[]
for key,value in models.items():
    value.fit(X_train,y_train)
    y_pred= value.predict(X_test)
    accuracy= accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)
    print(key)
    print(accuracy)
```

Since xgboost does not support categorical variables directly, it must be fed a label encoded y variable. So i train it separately. 

```python
X_xgb = df.drop('Stay',axis=1)
y_xgb = df['Stay']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_encoded = le.fit_transform(y_xgb)
```

```python
y_encoded
```

```python
X_train_xgb , X_test_xgb, y_train_xgb, y_test_xgb = train_test_split(X_xgb, y_encoded, test_size = 0.3,
                                                                    random_state = 42)
```

```python
xgb = XGBClassifier()
```

```python
xgb.fit(X_train_xgb,y_train_xgb)
```

```python
y_pred_xgb = xgb.predict(X_test_xgb)
```

```python
accuracy_xgb = accuracy_score(y_test_xgb, y_pred_xgb)
print(accuracy_xgb)
```

Since the scores are quite low, highest being 43% accuracy for catboost, followed by 42% for xgbclassifier, trying with df2 which has lesser number of output classes might make a difference.

## Training models using df2 (Output classes grouped)

```python
value_2 = [ RandomForestClassifier(), DecisionTreeClassifier(),  
        KNeighborsClassifier(), CatBoostClassifier()]

key_2 = [ 'RandomForsetClassifier', 'DecisionTreeClassifier',  
      'KNeighborsClassifier', 'CatBoostClassifier']

models_2 = dict(zip(key_2,value_2))
accuracy_scores_2=[]
for key,value in models_2.items():
    value.fit(X_train_2,y_train_2)
    y_pred_2= value.predict(X_test_2)
    accuracy_2= accuracy_score(y_test_2, y_pred_2)
    accuracy_scores_2.append(accuracy_2)
    print(key)
    print(accuracy_2)
```

```python
X_xgb_2 = df2.drop('Stay',axis=1)
y_xgb_2 = df2['Stay']
y_encoded_2 = le.fit_transform(y_xgb_2)
X_train_xgb_2 , X_test_xgb_2, y_train_xgb_2, y_test_xgb_2 = train_test_split(X_xgb_2, y_encoded_2, test_size = 0.3,
                                                                    random_state = 42)
xgb_2 = XGBClassifier()
xgb_2.fit(X_train_xgb_2,y_train_xgb_2)
```

```python
y_pred_xgb_2 = xgb_2.predict(X_test_xgb_2)
accuracy_xgb_2 = accuracy_score(y_test_xgb_2, y_pred_xgb_2)
print(accuracy_xgb_2)
```

It's clear that grouping the classes helped increase the accuracy a lot. The best models are therefore catboost and xgboost with accuracies of 53% each. This is not a high enough score to be useful for deployment. However, the project has demonstrated that grouping the output classes is a viable strategy for improving model performance.
